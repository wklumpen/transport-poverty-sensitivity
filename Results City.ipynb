```python
import pandas as pd
import altair as alt

region = {
    "BOS": "Boston",
    "CHI": "Chicago",
    "LA": "Los Angeles",
    "NYC": "New York City",
    "PHL": "Philadelphia",
    "SFO": "San Francisco Bay Area",
    "WAS": "Washington, DC"
}

demos = {
    "B03002_001E": "Everyone",
    "B03002_004E": "Black",
    "B03002_003E": "White",
    "B03002_006E": "Asian American/Pacific Islander",
    "B03002_012E": "Hispanic/Latino",
    "B11003_016E": "Single Mother",
    "age_65p": "Age 65+",
    "low_income": "In Poverty",
    "zero_car_hhld": "Zero-Car Household",
}

DEMOS_TO_INCLUDE = ["B03002_001E", "B03002_003E", "zero_car_hhld", "B03002_004E", "low_income"]

tod = {
    "WEDAM": "Wednesday Morning",
    "WEDPM": "Wednesday Evening",
    "SATAM": "Saturday Morning"
}

fraction = pd.read_csv("../data/fraction.csv")
fraction = fraction[fraction.group.isin(DEMOS_TO_INCLUDE)]
fraction["group"] = fraction["group"].map(demos)
fraction["region"] = fraction["region"].map(region)
fraction["tod"] = fraction["tod"].map(tod)

percentile = pd.read_csv("../data/percentile.csv")
percentile = percentile[percentile.group.isin(DEMOS_TO_INCLUDE)]
percentile["group"] = percentile["group"].map(demos)
percentile["region"] = percentile["region"].map(region)
percentile["tod"] = percentile["tod"].map(tod)

auto = pd.read_csv("../data/auto.csv")
auto = auto[auto.group.isin(DEMOS_TO_INCLUDE)]
auto["group"] = auto["group"].map(demos)
auto["region"] = auto["region"].map(region)
auto["tod"] = auto["tod"].map(tod)
```

```python
#Let's just do WEDAM
fraction["name"] = "Fraction of Total Jobs"
percentile["name"] = "Population-Weighted Percentile"
auto["name"] = "Ratio of Transit/Auto"
# print(percentile.tod.unique())
# all = pd.concat(
#     [
#         fraction[fraction.tod == "Wednesday Morning"], 
#         percentile[percentile.tod == "Wednesday Morning"], 
#         auto[auto.tod == "Wednesday Morning"],
#     ],
#     axis="index"
#     )

# all = all.drop(columns=["tod"])
alt.data_transformers.disable_max_rows()
to_plot = auto[(auto.tod == "Wednesday Morning") & (auto.region == "New York City") & (auto.group == "Everyone")].drop(columns=["tod"]).melt(id_vars=["group", "region", "p", "name"])
to_plot.variable = to_plot.variable.map({"fgt0": "0", "fgt1":"1", "fgt2": "2"})


alt.Chart(to_plot).mark_line(strokeWidth=2, color="black").encode(
    alt.X("p:Q", title="% of Auto Accessible Jobs Reachable by Transit"),
    alt.Y("value:Q", title="Level of Disadvantage").axis(titleFontWeight="normal", format="%"),
    # alt.Color("group:N", title="Population Group", sort=["Everyone"]),
    alt.StrokeDash("variable:N", title="Value of "),
    # alt.Row("region:N", title=None).header(labelFontWeight="bold", labelFontSize=12),
    # alt.Column("name:N", title=None,).header(labelFontSize=12, labelFontWeight="bold")
).properties(
    # title = {
    #     "text": "Fraction of Total Opportunities FGT Measures in San Francisco Bay Area",
    #     "subtitle": "Measure: jobs reachable in 45 minutes from 7-9am on Wednesday, March 6, 2024"
    # },
    width=300,
    height=200
).configure(
    font="Lato",
).configure_view(
    strokeWidth=0
).configure_axis(
    grid=False,
    labelFontSize=10,
    titleFontSize=12
).configure_title(
    fontSize=16,
    anchor="start"
)
```
